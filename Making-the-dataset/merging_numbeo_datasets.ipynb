```python
import pandas as pd
import os


base_path = r"D:\MaraBackup\Facultate\Master-sem2\AnalizaDatelor\numbeo\datasets"
folders = [
    "cost_of_living_index",
    "crime_index",
    "health_care_index",
    "pollution_index",
    "quality_of_life_index",
    "traffic_index"
]

columns_to_drop_before = {
    "cost_of_living_index": ["Rank","Cost of Living Plus Rent Index", "Local Purchasing Power Index"],
    "crime_index": ["Rank"],
    "health_care_index": ["Rank","Health CareExp. Index"],
    "pollution_index": ["Rank","Exp Pollution Index"],
    "quality_of_life_index": ["Rank","Purchasing Power Index", "Safety Index", "Health Care Index", "Cost of Living Index", "Property Price to Income Ratio",
                              "Traffic Commute Time Index", "Pollution Index"],
    "traffic_index": ["Rank","Time Exp. Index"]
}

ignored_folders = [
#     "quality_of_life_index",
#     "quality_of_life_index"
]

columns_to_drop_after = []

dfs = []
for folder in folders:
    
    if folder in ignored_folders:
        continue
    
    folder_path = os.path.join(base_path, folder)
    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            df = pd.read_csv(os.path.join(folder_path, file))

            
            for col in columns_to_drop_before.get(folder, []):
                if col in df.columns:
                    df.drop(columns=[col], inplace=True)

            
            df.set_index(df.columns[0], inplace=True)

            
            df = df.add_suffix("_" + folder)

            
            df.index.name = "Country"

            dfs.append(df)


from functools import reduce
merged_df = reduce(lambda left, right: pd.merge(left, right, on="Country", how="outer"), dfs)


merged_df.drop(columns=columns_to_drop_after, errors='ignore', inplace=True)
merged_df = merged_df.astype(str, errors='ignore')

output_path = os.path.join(base_path, "combined_numbeo_dataset_adjusted.csv")
merged_df.to_csv(output_path)

print(merged_df.shape[0])
merged_df.head()
```
