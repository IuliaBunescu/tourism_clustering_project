## Making the Final Dataset- Correlation between safety and average stays

```python
import pandas as pd
```

```python
numbeo_dataset = pd.read_csv("../Data/combined_numbeo_dataset_adjusted.csv")
tourism_dataset = pd.read_csv("../Data/tourism_combined_data.csv")
```

```python
combined_df_inner = pd.merge(tourism_dataset,numbeo_dataset, on="Country", how="inner")
combined_df_inner 
```

```python
combined_df_inner.drop(columns=["Rent Index_cost_of_living_index", "Crime Index_crime_index", "CO2Emission Index_traffic_index"], inplace=True)
```

```python
combined_df_inner.columns
```

```python
combined_df_inner["inverse_inefficiency_index"] = 1 / combined_df_inner["Inefficiency Index_traffic_index"]
combined_df_inner["inverse_commute_time_index"] = 1 / combined_df_inner["Time Index(in minutes)_traffic_index"]
combined_df_inner["inverse_traffic_index"] = 1 / combined_df_inner["Traffic Index_traffic_index"]
combined_df_inner["inverse_pollution_index"] = 1 / combined_df_inner["Pollution Index_pollution_index"]
```

```python
#We want to determine the correlation between average stay and safety
#before normalization
correlation = combined_df_inner['avg_stay_days_over_time'].corr(combined_df_inner['Safety Index_crime_index'])
print('Correlation between average stay days and safety is:', correlation)
```

```python
#let's normalize the data from the two columns since there are different ranges

# Drop rows where 'Safety Index_crime_index' has NaN values
combined_df_inner = combined_df_inner.dropna(subset=['Safety Index_crime_index'])
# Drop rows where 'avg_stay_days_over_time' has NaN values
combined_df_inner = combined_df_inner.dropna(subset=['avg_stay_days_over_time'])


columns_to_normalize = ['Safety Index_crime_index', 'avg_stay_days_over_time']
#normalize the two columns with min-max method
combined_df_inner[columns_to_normalize] = combined_df_inner[columns_to_normalize].apply(lambda x: (x-x.min())/(x.max()-x.min()))                                                    
correlation = combined_df_inner['avg_stay_days_over_time'].corr(combined_df_inner['Safety Index_crime_index'])
print('Correlation between average stay days and safety after data normalization is:', correlation)
```
